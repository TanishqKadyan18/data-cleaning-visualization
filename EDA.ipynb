```python
import pandas as pd
import numpy as np

# Set random seed for reproducibility
np.random.seed(42)

# Generate 100 rows of synthetic student performance data
study_hours = np.random.randint(1, 15, 100)        # hours studied per week
attendance = np.random.randint(40, 100, 100)       # attendance percentage
assignments = np.random.randint(1, 10, 100)        # assignments completed

# Rule for pass/fail (target variable)
pass_fail = [
    1 if (h > 6 and a > 70 and asg > 5) else 0
    for h, a, asg in zip(study_hours, attendance, assignments)
]

# Create DataFrame
df = pd.DataFrame({
    "StudyHours": study_hours,
    "Attendance": attendance,
    "Assignments": assignments,
    "Pass": pass_fail
})

# Show first 5 rows
df.head()
```

```python
df.to_csv("student_performance.csv", index=False)
print("CSV file created and saved in notebook!")
```

```python
import pandas as pd

# Load the dataset (use your student_performance.csv or any other dataset)
df = pd.read_csv("student_performance.csv")

# Quick overview
print(df.shape)
df.head()
```

```python
# Summary statistics
df.describe()

# Check for missing values
df.isnull().sum()

# Data types
df.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histogram of Study Hours
sns.histplot(df["StudyHours"], bins=10, kde=True)
plt.title("Distribution of Study Hours")
plt.show()

# Countplot of Pass/Fail
sns.countplot(x="Pass", data=df)
plt.title("Pass/Fail Distribution")
plt.show()
```

```python
# Study Hours vs Pass
sns.boxplot(x="Pass", y="StudyHours", data=df)
plt.title("Study Hours vs Pass")
plt.show()

# Attendance vs Pass
sns.boxplot(x="Pass", y="Attendance", data=df)
plt.title("Attendance vs Pass")
plt.show()
```
